# 3D graph trial

```python
###3d graph
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
fig=plt.figure()
ax=plt.axes(projection="3d")

num_bars=10
x_pos=random.sample(range(30),num_bars);print(x_pos)
y_pos=random.sample(range(30),num_bars);print(y_pos)
z_pos=[0]*num_bars;print(z_pos)
x_size=np.ones(num_bars);print(x_size)
y_size=np.ones(num_bars)
z_size=random.sample(range(30),num_bars);print(z_size)

ax.bar3d(x_pos,y_pos,z_pos,x_size,y_size,z_size,color='aqua')
plt.show()
```
